```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bikedata=pd.read_csv("SeoulBikeData.csv", encoding= 'unicode_escape')
```

```python
bikedata.head()
```

```python
bikedata=bikedata.drop('Date',axis=1)
```

```python
bikedata.head()
```

```python
bikedata.Seasons=pd.Categorical(bikedata.Seasons)
bikedata['season_code']=bikedata.Seasons.cat.codes
#changing seasons in string type to categorical 
bikedata=bikedata.drop('Seasons',axis=1)
```

```python
bikedata.Holiday=pd.Categorical(bikedata.Holiday)
bikedata['holiday_code']=bikedata.Holiday.cat.codes
#changing seasons in string type to categorical 
bikedata=bikedata.drop('Holiday',axis=1)
```

```python
bikedata['Functioning Day']=pd.Categorical(bikedata['Functioning Day'])
bikedata['functioning_day_code']=bikedata['Functioning Day'].cat.codes
#changing seasons in string type to categorical 
bikedata=bikedata.drop('Functioning Day',axis=1)
```

```python
# Checking for missing values

nullValues = pd.concat([bikedata.isnull().sum()],axis=1, keys=['Null Values'])
nullValues[nullValues.sum(axis=1) > 0 ]
```

### our dataset has no null values

```python
# Is any entries are duplicate or not?
print('Total Duplicate Entries: ' , bikedata.duplicated().sum())
print('Percentage of Duplication: ' , (bikedata.duplicated().sum() / len(bikedata) * 100).round(2) , '%')
```

### our data set has no duplicate values


```python
from sklearn.model_selection import train_test_split

bikedata
    
```

```python
x = bikedata.drop('Rented Bike Count',axis=1)
y = bikedata['Rented Bike Count']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15,random_state=0) #splitting 85% for training and 15% for testing
```

```python


from sklearn.ensemble import RandomForestRegressor 
#applying random forest classifier as predicting categorical values (language column)
#Gaussian Classifier
rfr=RandomForestRegressor(n_estimators=101, random_state=0)
rfr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_pred=rfr.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head() #observing few actual values and the predicted values
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))


```

```python
from sklearn.metrics import r2_score
```

```python
print("R2score",r2_score(y_test,y_pred)) #(coefficient of determination) regression score function.
```
